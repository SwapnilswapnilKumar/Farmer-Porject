```python
grouped = df.groupby('label')
```

```python
grouped[['temperature']].mean()
```

```python
grouped[['N']].mean()
```

```python
group=grouped['temperature'].agg(min='min',max='max',mean='mean')
group
```

```python
group.reset_index(inplace=True)
```

```python
x=list(group['label'])
y=list(group['mean'])
plt.figure(figsize=(14,7))
plt.bar(x,y,ec='k',color='maroon')
plt.title("avg temperature ")
plt.xlabel('label')
plt.ylabel('mean')
plt.xticks(rotation=90)
plt.show()
```

```python
x=list(group['label'])
y=list(group['min'])
plt.figure(figsize=(14,7))
plt.bar(x,y,ec='k',color='maroon')
plt.title("minimum temperature ")
plt.xlabel('label')
plt.ylabel('min')
plt.xticks(rotation=90)
plt.show()
```

```python
x=list(group['label'])
y=list(group['max'])
plt.figure(figsize=(14,7))
plt.bar(x,y,ec='k',color='maroon')
# plt.title("maximum temperature ")
plt.xlabel('label')
plt.ylabel('max')
plt.xticks(rotation=90)
plt.show()
```

```python
# normalizing outliers:->>>   # transformation
from sklearn.preprocessing import StandardScaler, MinMaxScaler
std = StandardScaler()
x_transformed = std.fit_transform(x) # standard normal distribution form
```

```python
print("total nuo. of clas samples: ",x_transformed.shape[0])
```

### training and testing splitting

```python
train = x_transformed[:2100]
test = x_transformed[2100:]
```

```python
#KMeans is a unsupervised ml algo
from sklearn.cluster import KMeans
obj = KMeans()
```

```python
obj.fit(train)
```

```python
clust = obj.labels_
```

```python
testing_df = x[:2100]
```

```python
testing_df['cluster'] = clust
```

```python
testing_df
```

```python
testing_df['Label'] = df[:2100]['label']
```

```python
testing_df.head(40)
```

```python
testing_df['cluster'].nunique()
```

```python
testing_df['cluster'].value_counts()
```

```python
fifth_cluster = testing_df[testing_df['cluster'] == 5]
```

```python
fifth_cluster.shape
```

```python
list(fifth_cluster['Label'].unique())
```

```python
obj.score(train,clust) ### training time
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(train,clust) *100
```
